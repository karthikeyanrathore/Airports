```python
import numpy
import pandas as pd
```

```python
flights = pd.read_csv("flightlist_20190101_20190131.csv")
print(flights.shape)
flights.head(10)
```

```python
print(flights.columns)
```

```python
flights = flights[["day" , "origin" , "destination"]]
flights = flights.dropna()
flights = flights.sample(1000 , random_state = 10)
print(flights.shape)
flights.head(10)
```

```python
flights.to_csv("flights_sample.csv", index= False)
```
